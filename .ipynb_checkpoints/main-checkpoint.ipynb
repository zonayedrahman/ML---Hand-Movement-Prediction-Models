```python
# Distribution graphs (histogram/bar graph) of column data
def plotPerColumnDistribution(df):
    plt.hist(df["Class"])
    plt.xlabel("Class")
    # plt.xlim([0, 2])
    plt.xticks([0, 1, 2])
    plt.show()
```

```python
# Scatter and density plots
def plotScatterMatrix(df):
    x = df1.iloc[:, 1:]
    # output
    y = df1.iloc[:, 0]

    fig, axs = plt.subplots(56, 2, figsize=(18,70))

    col_names = df.columns[1:]
    for i in range(0,56):
        for j in range(0,2):
            axs[i, j].scatter(x.iloc[:,i*2+j],y, s=1)
            axs[i, j].set_xlabel(col_names[i*2+j])
            axs[i, j].set_ylim([-1,3])

    for ax in axs.flat:
        ax.set( ylabel='Class')

    fig.tight_layout()
```

```python
def corrstats(df):
    corrMatrix = df.corrwith(df["Class"]).sort_values(ascending=False)
    corrMatrix.set_option('display.max_rows', None)
    print("Correlation Coefficients Matrix")
    print(corrMatrix)
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn import linear_model
from sklearn.metrics import confusion_matrix, accuracy_score


def basic_logistic_regression(df):
    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)

    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.25, random_state=42)

    log_reg = linear_model.LogisticRegression(max_iter=3000)
    log_model = log_reg.fit(xtrain, ytrain)

    y_pred = log_model.predict(xtest)

    cm = confusion_matrix(ytest, y_pred)
    print (f"Confusion Matrix of {df.name}: \n", cm)
    print ("Accuracy : ", accuracy_score(ytest, y_pred))
```

```python
df1.name = "user_a"
basic_logistic_regression(df1)
df2.name = "user_b"
basic_logistic_regression(df2)
df3.name = "user_c"
basic_logistic_regression(df3)
df4.name = "user_d"
basic_logistic_regression(df4)
```

```python
def poly_logistic_regression(df, degree):
    x = df.iloc[:, 1:].values
    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()
    x = scale.fit_transform(x)

    poly = PolynomialFeatures(degree=degree)
    x = poly.fit_transform(x)

    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.25, random_state=42)


    log_reg = linear_model.LogisticRegression(max_iter=3000)
    log_model = log_reg.fit(xtrain, ytrain)

    y_pred = log_model.predict(xtest)

    cm = confusion_matrix(ytest, y_pred)
    print (f"Confusion Matrix of {df.name} with feature degree of {degree}: \n", cm)
    print ("Accuracy : ", accuracy_score(ytest, y_pred))
```

```python
df1.name = "user_a"
poly_logistic_regression(df1, 2)
df2.name = "user_b"
poly_logistic_regression(df2, 2)
df3.name = "user_c"
poly_logistic_regression(df3, 2)
df4.name = "user_d"
poly_logistic_regression(df4, 2)
```

```python
def l1_k_fold_logistic_regression(df):
    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)


    poly = PolynomialFeatures(degree=2)
    x = poly.fit_transform(x)
    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)


    alphas = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0]
    # alphas = [1.0, 10.0, 100.0, 1000.0]
    best_alph = [0] * 9
    acc = [0] * 9
    print("L1 REGULARIZATION: ")
    for i in range(2, 11):
        cv = KFold(n_splits=i, random_state=42, shuffle=True)
        best_score = 0
        best_c = 0.0001
        for c in alphas:
            log_reg = linear_model.LogisticRegression(penalty='l1',C=c,solver="liblinear", max_iter=3000)
            log_model = log_reg.fit(xtrain, ytrain)
            scores = cross_val_score(log_reg, x, y, scoring='accuracy', cv=cv, n_jobs=-1)
            # print(f"scores with fold={i} and c={c}:", max(scores))
            if max(scores) > best_score:
                best_score = max(scores)
                best_c = c

        print("best C:", best_c, "Highest acc: ", best_score)
        best_alph[i-2] = best_c
        acc[i-2] = best_score



    print("END:", best_alph, acc)
    return best_alph, acc



```

```python
df1_l1Cs, df1_l1Accs = l1_k_fold_logistic_regression(df1)
```

```python
df2_l1Cs, df2_l1Accs = l1_k_fold_logistic_regression(df2)
```

```python
df3_l1Cs, df3_l1Accs = l1_k_fold_logistic_regression(df3)
```

```python
df4_l1Cs, df4_l1Accs = l1_k_fold_logistic_regression(df4)
```

```python
def l2_k_fold_logistic_regression(df):
    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)


    poly = PolynomialFeatures(degree=2)
    x = poly.fit_transform(x)
    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)


    alphas = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0]
    # alphas = [1.0, 10.0, 100.0, 1000.0]
    best_alph = [0] * 9
    acc = [0] * 9
    print("L2 REGULARIZATION: ")
    for i in range(2, 11):
        cv = KFold(n_splits=i, random_state=42, shuffle=True)
        best_score = 0
        best_c = 0.0001
        for c in alphas:
            log_reg = linear_model.LogisticRegression(penalty='l2',C=c,solver="liblinear", max_iter=3000)
            log_model = log_reg.fit(xtrain, ytrain)
            scores = cross_val_score(log_reg, x, y, scoring='accuracy', cv=cv, n_jobs=-1)
            # print(f"scores with fold={i} and c={c}:", max(scores))
            if max(scores) > best_score:
                best_score = max(scores)
                best_c = c

        print("For k=", i, "best C:", best_c, "Highest acc: ", best_score)
        best_alph[i-2] = best_c
        acc[i-2] = best_score



    print("END:", best_alph, acc)
    return best_alph, acc



```
