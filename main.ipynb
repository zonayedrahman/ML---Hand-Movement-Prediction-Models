```python
# Distribution graphs (histogram/bar graph) of column data
def plotPerColumnDistribution(df):
    plt.hist(df["Class"])
    plt.xlabel("Class")
    # plt.xlim([0, 2])
    plt.xticks([0, 1, 2])
    plt.show()
```
